```python
from transformers import AutoModel, AutoTokenizer
import torch

# Load model and tokenizer
model_name = "emanjavacas/GysBERT"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name, dtype="auto")

# Set to evaluation mode
model.eval()

# Example inference
text = "Dit is een testzin in het Nederlands."
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

# Run inference (no gradient computation for speed)
with torch.no_grad():
    outputs = model(**inputs)
    
# Get hidden states (last layer by default)
embeddings = outputs.last_hidden_state
# Pool to get sentence embedding (mean pooling)
sentence_embedding = embeddings.mean(dim=1).squeeze()

print(f"Input shape: {inputs['input_ids'].shape}")
print(f"Embedding shape: {sentence_embedding.shape}")
print(f"Sentence embedding (first 10 values): {sentence_embedding[:10]}")
```
